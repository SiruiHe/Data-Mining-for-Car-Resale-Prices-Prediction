```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('train.csv')
```

```python
# Distribution of Numerical Features
numerical_features = ['manufactured', 'no_of_owners', 'coe']

plt.figure(figsize=(10, 8))
for feature in numerical_features:
    df[feature] = df[feature].fillna(df[feature].median())
    df[feature] = (df[feature] - df[feature].min()) / (df[feature].max() - df[feature].min())
    plt.subplot(3, 5, numerical_features.index(feature) + 1)
    sns.histplot(data=df[feature], bins=10, kde=True)
    plt.title(feature)
plt.tight_layout()
plt.show()
```

```python
# Distribution of Numerical Features
numerical_features = ['curb_weight', 'power', 'engine_cap',
                      'depreciation', 'road_tax', 'dereg_value', 'mileage',
                      'omv', 'arf']

plt.figure(figsize=(10, 8))
for feature in numerical_features:
    df[feature] = df[feature].fillna(df[feature].median())
    df[feature + '_log'] = np.log(df[feature]+1)
    df[feature + '_log'] = (df[feature + '_log'] - df[feature + '_log'].min()) / (df[feature + '_log'].max() - df[feature + '_log'].min())
    plt.subplot(3, 5, numerical_features.index(feature) + 1)
    sns.histplot(data=df[feature + '_log'], bins=10, kde=True)
    plt.title(feature)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('train.csv')

numerical_features = ['curb_weight', 'power', 'engine_cap',
                      'depreciation', 'road_tax', 'dereg_value', 'mileage',
                      'omv', 'arf']

plt.figure(figsize=(10, 8))
for feature in numerical_features:
    df[feature] = df[feature].fillna(df[feature].median())
    
    df[feature + '_bin'] = pd.cut(df[feature], bins=10, labels=False)
    
    plt.subplot(3, 5, numerical_features.index(feature) + 1)
    sns.histplot(data=df[feature + '_bin'], bins=10, kde=False)
    plt.title(feature)
    
plt.tight_layout()
plt.show()
```
